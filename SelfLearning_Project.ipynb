```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression # Corrected typo 'modal' to 'model'
from sklearn.semi_supervised import SelfTrainingClassifier # Corrected casing 'Semi_supervised' to 'semi_supervised'
from sklearn.preprocessing import LabelEncoder
```

```python
detail=pd.read_csv('/content/Titanic Data.csv')
```

```python
detail=detail[["Survived","Pclass","Sex","Age","Fare"]].dropna()
```

```python
le=LabelEncoder()
detail["Sex"]=le.fit_transform(detail["Sex"])
```

```python
np.random.seed(42)
mask = np.random.rand(len(detail)) < 0.5
detail.loc[mask, "Survived"] = -1
```

```python
X = detail.drop("Survived", axis=1)
Y = detail["Survived"]
```

```python
base_model=LogisticRegression()
```

```python
self_training_model=SelfTrainingClassifier(base_model,criterion ='k_best')
```

```python
self_training_model.fit(X,Y)
```

```python
detail["Predicted_Label"] = self_training_model.predict(X)
```
